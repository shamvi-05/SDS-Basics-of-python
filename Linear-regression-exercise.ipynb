```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.title('canada per capita income')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='+')
```

```python
rgr=linear_model.LinearRegression()
```

```python
rgr.fit(df[['year']],df['per capita income (US$)'])
```

```python
rgr.predict([[2020]])
```

From the above linear regression model, predicted per capita income for year 2020 is approximately equal to 41288.694 in (US$) currency
